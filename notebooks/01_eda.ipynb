## Target Variable Observations

- TARGET = 0 indicates loan repayment.
- TARGET = 1 indicates loan default.
- The dataset is highly imbalanced.
- Approximately 92% of customers repaid their loans.
- Approximately 8% of customers defaulted.

```python
plt.figure(figsize=(6, 4))

sns.countplot(
    x="TARGET",
    data=df
)

plt.title("Loan Repayment Distribution")
plt.xlabel("Target")
plt.ylabel("Count")

plt.show()
```

## Class Distribution Observations

- The number of loan repayments is significantly higher than defaults.
- The imbalance may affect model performance.
- Accuracy alone will not be a reliable evaluation metric.
